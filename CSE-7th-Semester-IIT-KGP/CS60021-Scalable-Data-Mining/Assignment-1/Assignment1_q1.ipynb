```python
!pip install utility
import argparse
import time
import random
import numpy as np
import utility
import pandas as pd
import matplotlib.pyplot as plt

filename = 'Dataset_A.csv'
label = 'Segmentation'


def handle_missing_values(dataframe):
    """Function to read data from file
    Args:
        dataframe (Dataframe) : input dataframe

    Returns:
        dataframe: data frame with missing values replaced with MODE of that column
    """

    for col_name in dataframe.columns:
        dataframe[col_name].fillna(dataframe[col_name].mode()[0], inplace=True)
    return dataframe


def read_data():
    data = pd.read_csv(filename)
    return data


def prepare_data(data):
    dataclasses = []
    for i in range(1, len(data[label]) + 1):
        dataclasses.append(
            {
                'Gender': data['Gender'][i],
                'Ever_Married': data['Ever_Marries'][i],
                'Age': data['Age'][i],
                'Graduated': data['Graduated'][i],
                'Profession': data['Profession'][i],
                'Work_Experience': data['Work_Experience'][i],
                'Spending_Score': data['Spending_Score'][i],
                'Family_Size': data['Family_Size'][i],
                'Var_1': data['Var_1'][i]
            }
        )
    return dataclasses


def train_valid_test_split(data):
    random.shuffle(data)
    X_train, X_valid = data[:int(0.75 * len(data))], data[int(0.75 * len(data)):]
    return X_valid, X_train


def train_test_split(df, frac1=0.8, frac2=0.2):
    if abs((frac1 + frac2) - 1.0) > 1e-6:
        raise ValueError('Fractions should add up to 1')
    df = handle_missing_values(df)
    dataset = prepare_data(df)
    random.shuffle(dataset)
    X_train, X_test = dataset[:int((frac1 - frac2) * len(dataset))], dataset[int(frac1 * len(dataset)):]
    return X_test, X_train


def predict_one(decision_tree, data):
    if decision_tree.left is None and decision_tree.right is None:
        return decision_tree.mean
        # based on the decision either recurse to the left
        # or the right half until the leaf node is reached
    if data[decision_tree.attr] <= decision_tree.split:
        return predict_one(decision_tree.left, data)
    return predict_one(decision_tree.right, data)


def predict(decision_tree, data):
    mse, preds = 0, []
    for i in data:
        # make prediction for the current data point
        val = predict_one(decision_tree, i)
        # insert the prediction in the prediction array
        preds.append(val)
        mse += (val - i['Increase rate']) ** 2

    # calculate the average error
    mse = mse / len(data)
    return mse,


def r2_score(X, X_hat):
    ssr, sst = 0, 0
    Y_bar = sum(X) / len(X)
    for i, val in enumerate(X):
        ssr += (val - X_hat[i]) ** 2
        sst += (val - Y_bar) ** 2
    if sst == 0: sst = 10 ** -20
    score = 1 - ssr / sst
    return score


def get_accuracy(decision_tree, X_data):
    _, preds = predict(decision_tree, X_data)
    data = []
    for i in X_data:
        data.append(i['Increase rate'])
    acc = r2_score(data, preds)
    return acc


class Node:

    def __int__(self, attribute=None, value=None, mean=0.0, mse=0.0):
        """
                Constructor
        """
        # to store value of leaf node
        self.value = value

        # for decision node
        # self.threshold = threshold
        # self.feature_index = feature_index
        self.left = None
        self.right = None
        # self.information_gain = information_gain
        self.attribute = attribute
        self.mean = mean
        self.mse = mse

    def number_of_node(self):
        number = 1
        if self.left is not None:
            number += self.left.number_of_node()
        if self.right is not None:
            number += self.right.number_of_node()
        return number

    def restore_node(self, attr, left, right):
        self.left = left
        self.right = right
        self.attribute = attr

    def delete_node(self):
        self.left = None
        self.right = None
        self.attribute = label

def construct_tree(samples, current_height, max_height, attr_list):
    if current_height == max_height or len(samples) == 0:
        return None

        # if we have one data only then store it in the node and
        # consider this as the prediction alue for this node also
    if len(samples) == 1:
        return Node(label, samples[0][label], samples[0][label], 0)

        # select the best attribute to be selected for this node
        # with the help of the utility functions
    attr, split, mse = utility.good_attr(samples, attr_list)
    samples1, samples2, mean = [], [], 0

    for i in samples:
        mean += i[label]
        if i[attr] <= split:
            samples1.append(i)
        else:
            samples2.append(i)

    mean /= len(samples)
    head = Node(attr, split, mean, mse)
    head.left = construct_tree(
        samples1, current_height + 1, max_height, attr_list)
    head.right = construct_tree(
        samples2, current_height + 1, max_height, attr_list)
    if head.left == None and head.right == None:
        head.attr = 'Increase rate'
        head.split = head.mean

    return head


def select_best_tree_10(data, max_height, X_test):
    if max_height == -1:
        max_height = 300
    attributes = ['Gender', 'Ever_Married', 'Age', 'Graduated', 'Profession', 'Work_Experience', 'Spending_Score',
                  'Family_Size', 'Var_1', 'Segmentation']
    min_mse, tree, mse_avg, acc_avg = 10 ** 18, None, 0, 0
    train, valid = None, None
    for _ in range(10):
        X_valid, X_train = train_valid_test_split(data)
        decision_tree = construct_tree(X_train, 0, max_height, attributes)
        test_mse = get_accuracy(decision_tree, X_test)
        test_acc = predict(decision_tree, X_test)

        if test_mse < min_mse:
            min_mse = test_mse
            tree = decision_tree
            mse_avg += test_mse
            acc_avg += test_acc

    mse_avg = mse_avg / 10
    acc_avg = acc_avg / 10
    return tree, train, valid, mse_avg, acc_avg


if __name__ == '__main__':
    start = time.time()
    parser = argparse.ArgumentParser()
    parser.add_argument("--Height", help="Provide Model with the Maximum Height of Decision Tree for Question 1.",
                        type=int)

    args = parser.parse_args()
    height = -1
    if args.Height:
        height = args.Height

    print("\n ============= READING DATA ============ \n")
    df = read_data()
    print("\n =============  DATA FRAME  ============ \n")
    print(df)
    print("Time elapsed  =  {} ms".format(time.time() - start))
    print("\n ============= DATA READ ============ \n\n")
    test_feature, train_feature = train_test_split(df, frac1=0.8, frac2=0.2)
    print("\n========= TRAINING STARTED =========\n")

    X_train = train_feature
    start = time.time()
    tree, train, valid, mse_avg, acc_avg = select_best_tree_10(train_feature, height, test_feature)
    print("Time elapsed  =  {} ms".format(time.time() - start))
    print("\n ============= TRAINING FINISHED ============ \n")
    print("Average Test Accuracy: {}\n".format(acc_avg * 100))
    print("Average Test MSE: {}\n".format(mse_avg))

    # data_print(tree, train, test, valid)
    train = X_train
```
